```python
# ---------------------------------------------------------
# SECTION 4: DATA CLEANING & QUALITY PIPELINE
# Aim: Standardize data types and handle missing values.
# ---------------------------------------------------------

# Create a copy for cleaning to preserve raw data integrity
df = raw_df.copy()

# 1. Convert timestamp strings to actual datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 2. Extracting Temporal Features for later trend analysis
df['interaction_date'] = df['timestamp'].dt.date
df['hour_of_day'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.day_name()
df['month'] = df['timestamp'].dt.month_name()

# 3. Handling Missing Values in 'user_feedback_rating'
# Strategy: Median imputation to avoid outlier distortion
median_rating = df['user_feedback_rating'].median()
df['user_feedback_rating'].fillna(median_rating, inplace=True)
print(f"Action: Imputed missing ratings with median value: {median_rating}")

# 4. Handling Missing Values in 'estimated_business_impact'
# Strategy: Impute with 'Unknown' to maintain data distribution
df['estimated_business_impact'].fillna('Unknown', inplace=True)
print("Action: Imputed missing business impact with 'Unknown'.")

# 5. Outlier Detection in response_time_ms
# Using IQR method to identify anomalies
Q1 = df['response_time_ms'].quantile(0.25)
Q3 = df['response_time_ms'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['response_time_ms'] < lower_bound) | (df['response_time_ms'] > upper_bound)]
print(f"Quality Check: Identified {len(outliers)} latency outliers outside [{lower_bound}, {upper_bound}] ms.")

# 6. Type Normalization
# Ensuring user_feedback_rating is an integer for categorical grouping
df['user_feedback_rating'] = df['user_feedback_rating'].astype(int)

# 7. Final Verification of Cleaning Steps
print("\n--- Post-Cleaning Statistics ---")
print(df.isnull().sum())

# Logging logic for data flow validation
log_progress("Cleaning & Transformation")
# Final visual check of the cleaned dataframe structure
display(df.head(3))
```

```python
# ---------------------------------------------------------
# SECTION 9: ADVANCED FEATURE ENGINEERING
# Aim: Create informative features to support predictive models.
# ---------------------------------------------------------

# 1. Feature: Query Complexity Proxy
# Length of the query string as a proxy for complexity
df['query_length'] = df['user_query'].apply(len)

# 2. Feature: Efficiency Score
# High confidence in low time = High Efficiency
df['efficiency_score'] = df['bot_response_confidence'] / (df['response_time_ms'] / 1000)

# 3. Categorical Encoding
le = LabelEncoder()
categorical_cols = ['department', 'user_role', 'query_category', 'metrics_requested', 'analysis_type', 'estimated_business_impact']
for col in categorical_cols:
    df[f'{col}_encoded'] = le.fit_transform(df[col].astype(str))

# 4. Binary Indicator: High Impact
df['is_high_impact'] = df['estimated_business_impact'].apply(lambda x: 1 if x == 'High' else 0)

# 5. Segmenting Latency into buckets
df['latency_category'] = pd.qcut(df['response_time_ms'], q=3, labels=['Fast', 'Medium', 'Slow'])

print("Successfully engineered 5+ new features for predictive analysis.")
display(df[['interaction_id', 'efficiency_score', 'query_length', 'latency_category']].head())

log_progress("Feature Engineering")
```

```python
# ---------------------------------------------------------
# SECTION 10: PREDICTIVE MODELING EXPERIMENTS
# Aim: Train a Random Forest model to predict user ratings.
# ---------------------------------------------------------

# Define features and target
features = ['bot_response_confidence', 'response_time_ms', 'query_length', 
            'department_encoded', 'user_role_encoded', 'query_category_encoded',
            'analysis_type_encoded']
X = df[features]
y = df['user_feedback_rating']

# Splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_SEED)

# Scaling numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Model Training: Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=RANDOM_SEED)
rf_model.fit(X_train_scaled, y_train)

# Predictions and Metrics
y_pred = rf_model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Model Performance Metrics:")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared Score: {r2:.4f}")

# Feature Importance Visualization
importances = rf_model.feature_importances_
feat_importance = pd.Series(importances, index=features).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
feat_importance.plot(kind='barh', color='darkgreen')
plt.title("Feature Importance: Drivers of User Satisfaction")
plt.show()

log_progress("Predictive Modeling")
```
